```python
#----- seccion de bibliotecas

import import_ipynb
from ArbolesBinarios import AB
from ArbolesBinarios import Nodo

# -----


class ABB(AB):
  ''' 
  Clase arbol binario de busqueda. Genera la estructura de un arbol binario
  de busqueda asi como sus funciones.
  '''
  
  def __init__(self, dato:float):
    '''
    Metodo constructor de la clase ABB
    inputs:
        dato: valor del nodo raiz
    outputs:
        objeto ABB

    >>>ABB(170711)
    <__main__.ABB at ...>
    '''  
      
    # Llamamos al constructor de la clase padre con la palabra super
    super(ABB, self).__init__(dato)
    # Se agrega la variable ultimo agregado que sirve para clases posteriores
    self.ultimo_agregado = self.raiz

  def busqueda(self, nodo:Nodo, dato:float) -> Nodo:
    '''
    Metodo que busca un dato en el arbol y devuelve el nodo
    inputs:
        nodo: objeto de la clase Nodo a partir del que se iniciara la busqueda del dato
        dato: elemento a buscar
    outputs:
        Nodo: objeto de la clase Nodo en el que se encontro el dato

    >>>ABB(15).busqueda(ABB(15).raiz,15)
    <ArbolesBinarios.Nodo at 0x7fa1225bc090>
    '''   
      
    # si la raiz es None o el dato esta contenido en el nodo,
    # se devuelve el nodo.
    if nodo is None or nodo.dato == dato:
      return nodo
    # si el dato es mayor entonces se busca en el lado derecho
    if nodo.dato < dato:
      return self.busqueda(nodo.der, dato)
    # si no, se busca en el lado izquierdo
    else:
      return self.busqueda(nodo.izq, dato)

  def insertar_nodo(self, nodo:Nodo, dato:float) -> Nodo:
    '''
    Metodo que inserta un elemento en el arbol a partir de un nodo
    inputs:
        nodo: objeto de la clase Nodo a partir del que se insertara el dato
        dato: elemento a insertar
    outputs:
        Nodo: objeto de la clase Nodo donde se agrego el dato

    >>>ABB(170711).insertar_nodo(ABB(170711).raiz,13)
    <ArbolesBinarios.Nodo at 0x7fa129376c50>
    '''   
      
    # si el nodo es vacio ahi se crea el nuevo nodo
    if nodo is None:
      nuevo_nodo = Nodo(dato)
      self.ultimo_agregado = nuevo_nodo
      return nuevo_nodo
    # si el dato es menor que su padre, se inserta en el lado izquierdo
    if dato < nodo.dato:
      nuevo_nodo = self.insertar_nodo(nodo.izq, dato)
      nodo.izq = nuevo_nodo
      nuevo_nodo.padre = nodo
    # de no ser asi se inserta del lado derecho
    else:
      nuevo_nodo = self.insertar_nodo(nodo.der, dato)
      nodo.der = nuevo_nodo
      nuevo_nodo.padre = nodo
    # nodo guarda toda la ruta de donde sera insertado el dato
    # hasta caer en el caso base, es por eso que se devuelve    
    return nodo
  
  def insertar(self, dato:float):
    '''
    Metodo que inserta un nodo en el arbol, a partir de la raiz
    inputs:
        dato: elemento a insertar
    outputs:
        ----

    >>>ABB(15).insertar(170711)
    < >
    ''' 
      
    # si el dato a insertar no se encuentra en el ABB se inserta
    if self.busqueda(self.raiz, dato) is None:
      # Se inserta el dato desde la raiz
      self.insertar_nodo(self.raiz, dato)

  def minimo_en_subarbol(self, nodo:Nodo) -> Nodo:
    '''
    Metodo que busca el minimo en el subarbol a partir de nodo. Sirve
    para poder eliminar nodos del arbol y mantener el orden
    inputs:
        nodo: objeto de la clase Nodo en cuyo subarbol se buscara el minimo
    outputs:
        Nodo: objeto de la clase Nodo donde se encontro al minimo

    >>>arbol = ABB(1)
       arbol.insertar_nodo(arbol.raiz,13)
       arbol.minimo_en_subarbol(arbol.raiz)
    <ArbolesBinarios.Nodo at 0x7fa122505810>
    ''' 
      
    # caso base en el que no hay minimo
    if nodo is None:
      return nodo
    # caso base en el que el nodo es el minimo ya que no tiene hijo izq
    if nodo.izq is None:
      return nodo
    # llamada recursiva para movernos al siguiente minimo
    return self.minimo_en_subarbol(nodo.izq)


  def borra_nodo(self, nodo:Nodo, dato:float):
    '''
    Metodo que borra un nodo en el arbol, busca al nodo que contiene a dato y en caso de 
    existir lo borra; se pueden dar multiples casos
    inputs:
        nodo: objeto de la clase Nodo a borrar
        dato: elemento cuyo nodo se va a borrar (en caso de existir en el arbol)
    outputs:
        ---

    >>>p = ABB(170711)
       p.borra_nodo(p.raiz,170711) 
    < >
    ''' 
    
    # Caso 0) revisamos si el arbol es vacio.
    if self.raiz is None:
      return None
    # buscamos el nodo a borrar
    nodo_a_borrar = self.busqueda(nodo, dato)
    aux = nodo_a_borrar

    # Caso 0.1) si el dato no se encontro en el arbol no se puede borrar
    if nodo_a_borrar is None:
      return None

    # Caso 1) si el nodo a borrar es la RAIZ
    if nodo_a_borrar is self.raiz:
      # caso1.1) no tiene hijos, solo se borra la raiz
      if nodo_a_borrar.izq is None and nodo_a_borrar.der is None:
        self.raiz = None
        self.ultimo_agregado = None
        return None

      # Caso1.2) solo se tiene hijo derecho, entonces se sube al hijo derecho
      if nodo_a_borrar.izq is None and nodo_a_borrar.der is not None:
        self.raiz = nodo_a_borrar.der
        self.raiz.padre = None
        return self.raiz

      # Caso1.3) solo se tiene hijo izquierdo, entonces se sube al hijo izquierdo
      if nodo_a_borrar.izq is not None and nodo_a_borrar.der is None:
        self.raiz = nodo_a_borrar.izq
        self.raiz.padre = None
        return self.raiz

      # Caso1.4) tiene ambos hijos
      if nodo_a_borrar.izq is not None and nodo_a_borrar.der is not None:
        # buscamos el minimo en el subarbol derecho (minimo de los mayores)
        minimo = self.minimo_en_subarbol(nodo_a_borrar.der)
        aux = minimo.padre
        self.raiz.dato = minimo.dato
        self.borra_nodo(minimo, minimo.dato)
        return aux

    else: #Caso 2)
      # a partir de aqui se tienen 3 casos:
      # si no tiene hijos simplemente se borra el nodo
      # si tiene un solo hijo (ya sea izquierdo o derecho) se sube al unico hijo
      # tiene ambos hijos

      # es necesario identificar si el nodo a borrar es hijo izquierdo o derecho
      es_izquierdo = False
      if nodo_a_borrar.padre.izq == nodo_a_borrar:
        es_izquierdo = True

      # Caso2.1) no tiene hijos, solo se borra el nodo
      if nodo_a_borrar.izq is None and nodo_a_borrar.der is None:
        aux = nodo_a_borrar.padre
        # revisamos si el nodo a borrar es un hijo izquiero o derecho
        if es_izquierdo: # Caso2.1.1)
          aux.izq = None
        else: # Caso2.1.2)
          aux.der = None
        nodo_a_borrar = None
        return aux
      
      # Caso2.2) solo se tiene hijo izquierdo, entonces se sube al hijo izquierdo
      if nodo_a_borrar.izq is not None and nodo_a_borrar.der is None:
        nodo_a_borrar.izq.padre = nodo_a_borrar.padre
        aux = nodo_a_borrar.padre
        # revisamos si el nodo a borrar es un hijo izquiero o derecho
        if es_izquierdo: # Caso2.2.1)
          nodo_a_borrar.padre.izq = nodo_a_borrar.izq
        else: # Caso2.2.2)
          nodo_a_borrar.padre.der = nodo_a_borrar.izq
        return aux

      # Caso2.3) solo se tiene hijo derecho, entonces se sube al hijo derecho
      if nodo_a_borrar.izq is None and nodo_a_borrar.der is not None:
        nodo_a_borrar.der.padre = nodo_a_borrar.padre
        aux = nodo_a_borrar.padre
        # revisamos si el nodo a borrar es un hijo izquiero o derecho
        if es_izquierdo: # Caso2.3.1)
          nodo_a_borrar.padre.izq = nodo_a_borrar.der
        else: # Caso2.3.2)
          nodo_a_borrar.padre.der = nodo_a_borrar.der
        return aux

      # Caso2.4) tiene ambos hijos
      if nodo_a_borrar.izq is not None and nodo_a_borrar.der is not None:
        # buscamos el minimo en el subarbol derecho
        minimo = self.minimo_en_subarbol(nodo_a_borrar.der)
        aux = minimo.padre
        nodo_a_borrar.dato = minimo.dato
        self.borra_nodo(minimo, minimo.dato)
        return aux
  
  def borrar(self, dato:float):
    '''
    Metodo que borra un nodo que contiene a dato a partir de la raiz
    inputs:
        dato: elemento a borar
    outputs:
        ---

    >>>ABB(170711).borrar(170711) 
    < >
    ''' 
    self.borra_nodo(self.raiz, dato)

```

### Análisis

Indica cuál es el peor caso para este tipo de árboles y explica a qué orden de complejidad pertenecen las operaciones: eliminar árbol, buscar, insertar.

*Respecto a la complejidad en número de operaciones:*

**Peor caso**

El peor caso para este tipo de árbol ocurre cuando este es degenerado. Es decir, cuando el ABB luce como una lista. En esta situación, como revisamos ya en el parcial anterior; las operaciones efectuadas alcanzan el orden lineal, es decir: $O(n)$, donde $n$ denota el número de nodos que posee el árbol.


**Eliminar árbol**

Esta operación es de orden constante; es decir: $O(1)$, pues una vez que se elimina al nodo raíz del árbol, se libera todo el espacio en memoria asociado a éste. Y por tanto, ya no queda operación que realizar.

Por lo tanto, con una única operación se elimina todo el árbol.

**Buscar**

Al ser un árbol ordenado, se puede ver que en cada consulta a cada nodo, **se descarta a la mitad de nodos faltantes por comparar** (con el valor buscado). Esta diferencia respecto al Árbol Binario implica que en la mayoría de los casos, la operación de búsqueda sobre un ABB se mantenga en el orden de complejidad logarítmica, es decir: $O(log_2n)$ (donde $n$ denota el número de nodos que posee el ABB).

**Insertar**

Para poder insertar un elemento en el árbol se encesita primero localizar el nodo en que será insertado. Por el párrafo previo, sabemos que esta es una operación de complejidad logarítmica, y la *inserción* como tal del elemento solicitado es de orden constante. 

Por lo tanto, esta operación es de orden:

$$O(log_2n) + b,$$ 

con $b$ constante. 

Por cálculo, sabemos que nos interesa acotar la función que presenta mayor crecimiento; en este caso: $O(log_2n)$. 

Por tanto, el orden de complejidad al que pertenece esta operación es logarítmico.

2. Eliminar el valor 9, explica cada paso de manera detallada que se requiere para eliminar este valor.

    **Solución.** 

    Primero busca el nodo en que se encuentra el número 9 de la siguiente manera:

    **I.** Compara el 9 con el valor de la raíz (10), como 10 > 9, entonces se dirije a continuar la búsqueda en el subárbol de su hijo izquierdo. 

    **II.** Compara el 9 con el valor de este hijo izquierdo (5), como 5 < 9, entonces continúa la búsqueda en el hijo derecho de este nodo. 

    **III.** Finalmente compara el valor de este hijo derecho con el 9. Como `9 = 9`; ya ha encontrado el elemento a borrar dentro del ArbolAVL.

    **IV.** Se borra el nodo objetivo, pues es una hoja (no tiene hijo alguno).

    **V.** Ahora se debe asegurar que el árbol permanece balanceado (obteniendo los factores de equilibrio). Primero, el nodo que contiene al 5 está balanceado, pues si $H_I$ y $H_D$ denotan a sus hijos derecho e izquierdo, respectivamente, se tiene que: 
    
  $$|altura(H_I) - altura(H_D)|=0 \leq 1$$

    **VI.** Luego, para verificar que la raiz esta balanceda; se da cuenta que se cumple el caso donde **FE**(nodo) = -2 y (**FE**(nodo.der)=-1, entonces se realiza una rotación simple a la izquierda, colocando al hijo derecho de la raíz original como la nueva raíz.

    **VII.** Finalmente, el árbol se encuentra ya correctamente balanceado y se da por terminado el proceso.

```python
import import_ipynb
from ArbolesBinariosBusqueda import ABB
from ArbolesBinarios import Nodo

class ArbolAVL(ABB):

  def __init__(self, dato:float):
    '''
    Metodo constructor que hace uso del constructor de la clase padre (ABB)
    inputs:
        dato: elemento que ira en la raiz
    outputs:
        nodo: objeto de la clase ArbolAVL

    >>> ArbolAVL(170711) 
    <__main__.ArbolAVL at ....> 
    ''' 
    super(ArbolAVL, self).__init__(dato)
      
  def ultimo(self) -> Nodo:
    '''
    Metodo que devuelve el ultimo nodo agregado para balanceo en caso necesario
    inputs:
        ---
    outputs:
        Nodo: el ultimo nodo agregado 

    >>> ArbolAVL(11).ultimo() 
    <ArbolesBinarios.Nodo at 0x7fa1223b9710>
    ''' 
    return self.ultimo_agregado

  def balanceado(self, nodo:Nodo) -> bool:
    '''
    Metodo que determina si un nodo esta balanceado
    inputs:
        nodo; el cual se busca saber si esta balanceado
    outputs:
        Nodo: el ultimo nodo agregado 

    >>> ArbolAVL(11).balanceado(ArbolAVL(11).raiz) 
    < True > 
    ''' 
    return abs(self.altura(nodo.der)-self.altura(nodo.izq)) <= 1

  '''
   2 = el subarbol esta muy cargado a la derecha
   1 = el subarbol esta cargado a la derecha
   0 = no hace falta rotar
  -1 = el subarbol esta cargado a la derecha
  -2 = el subarbol esta muy cargado a la izquierda
  '''
  def factor_equilibrio(self, nodo:Nodo) -> int:
    '''
    Metodo que devuelve el factor de equilibrio de un nodo
    inputs:
        nodo: objeto de la clase Nodo; el cual se busca saber si esta balanceado
    outputs:
        int: factor de equilibrio

    >>> ArbolAVL(170711).factor_equilibrio(ArbolAVL(170711).raiz) 
    < 0 > 
    ''' 
    if nodo is None:
      return 0
    else:
      return self.altura(nodo.der)-self.altura(nodo.izq)
  
  def balancea(self, nodo:Nodo):
    '''
    Metodo que balancea el arbol, a partir de un nodo
    inputs:
        nodo: objeto de la clase Nodo; a partir del cual se balanceara el arbol
    outputs:
        ---

    >>> ArbolAVL(170711).balancea(ArbolAVL(170711).raiz) 
    < > 
    ''' 
    #caso base
    if nodo is None:
      return
    if not self.balanceado(nodo):
      factorNodo = self.factor_equilibrio(nodo)
      factorIz = self.factor_equilibrio(nodo.izq)
      factorDer = self.factor_equilibrio(nodo.der)
      #Si el nodo esta desbalanceado se tendra uno de 4 casos
      #mismos que se solucionan con alguna de las 4 rotaciones
      if factorNodo == -2 and (factorIz == -1 or factorIz == 0): #probado
        self.rotacionSimpleDerecha(nodo)
      if factorNodo == 2 and (factorDer == 1 or factorDer == 0): #probado
        self.rotacionSimpleIzquierda(nodo)
      if factorNodo == -2 and factorIz == 1: #probado
        self.rotacionIzquierdaDerecha(nodo)
      if factorNodo == 2 and factorDer == -1: #probado
        self.rotacionDerechaIzquierda(nodo) 
    #llamada recursiva para balancear toda la ruta. Desde donde se inserto
    #el nodo, hasta la raiz        
    self.balancea(nodo.padre)        
  
  def rotacionSimpleDerecha(self, nodo:Nodo):
    """
    Metodo que rota el nodo a la derecha
    inputs:
        nodo: objeto de la clase Nodo a rotar a la derecha
    outputs:
        ---

    """
    if nodo is not None:
      if nodo is self.raiz:
        #actualizamos la raiz
        self.raiz = nodo.izq
        self.raiz.padre = None
        # si el nodo.izq tiene hijo derecho se pasa como izq de nodo
        if nodo.izq.der is not None:
          nodo.izq.der.padre = nodo
          nodo.izq = nodo.izq.der
        else:
          nodo.izq = None
        self.raiz.der = nodo
        nodo.padre = self.raiz
      else:
        nuevaraiz = nodo.izq
        #actualizamos la nueva raiz
        if nodo.esIzq():
          nodo.padre.izq = nuevaraiz
        else:
          nodo.padre.der = nuevaraiz
        nuevaraiz.padre = nodo.padre
        # si el nodo.izq tiene hijo derecho se pasa como izq de nodo
        if nuevaraiz.der is not None:
          nuevaraiz.der.padre = nodo
          nodo.izq = nuevaraiz.der
        else:
          nodo.izq = None
        nuevaraiz.der = nodo
        nodo.padre = nuevaraiz
              
  def rotacionSimpleIzquierda(self, nodo:Nodo):
    """
    Metodo que rota el nodo a la izquierda
    inputs:
        nodo: objeto de la clase Nodo a rotar a la izquierda
    outputs:
        ---
        
    """
    if nodo is not None:
      if nodo is self.raiz:
        #actualizamos la raiz
        self.raiz = nodo.der
        self.raiz.padre = None
        # si el nodo.der tiene hijo izquierdo se pasa como der de nodo
        if nodo.der.izq is not None:
          nodo.der.izq.padre = nodo
          nodo.der = nodo.der.izq
        else:
          nodo.der = None
        self.raiz.izq = nodo
        nodo.padre = self.raiz
      else:
        nuevaraiz = nodo.der
        #actualizamos la nueva raiz
        if nodo.esIzq():
          nodo.padre.izq = nuevaraiz
        else:
          nodo.padre.der = nuevaraiz
        nuevaraiz.padre = nodo.padre
        # si el nodo.der tiene hijo izquierdo se pasa como der de nodo
        if nuevaraiz.izq is not None:
          nuevaraiz.izq.padre = nodo
          nodo.der = nuevaraiz.izq
        else:
          nodo.der = None
        nuevaraiz.izq = nodo
        nodo.padre = nuevaraiz
                      
  def rotacionIzquierdaDerecha(self, nodo:Nodo):
    """
    Metodo que define la doble rotacion a la derecha, a partir de un nodo    
    inputs:
        nodo: objeto de la clase Nodo a partir del cual se definira la doble rotacion a la derecha
    outputs:
        ---
        
    """
    self.rotacionSimpleIzquierda(nodo.izq)
    self.rotacionSimpleDerecha(nodo)
   
  def rotacionDerechaIzquierda(self, nodo:Nodo):
    """
    Metodo que define la doble rotacion a la izquierda, a partir de un nodo
    inputs:
        nodo: objeto de la clase Nodo a partir del cual se definira la doble rotacion a la izquierda
    outputs:
        ---
        
    """
    self.rotacionSimpleDerecha(nodo.der)
    self.rotacionSimpleIzquierda(nodo)

  def insertar(self, dato:float):
    """
    Metodo que inserta un nodo en el arbol y lo balancea, si es que se desbalanceo
    inputs:
        dato: elemento a insertar
    outputs:
        ---

    >>>ArbolAVL(170711).insertar(8) 
    <  >
    """
      
    #Se inserta el dato usando el metodo de la clase padre
    super(ArbolAVL, self).insertar(dato)
    #Identificamos donde se inserto el dato
    ultimo_nodo_insertado = self.ultimo()
      
    #Se rebalancean los nodos que hayan quedado desbalanceados
    #Y se empieza desde el padre, por que nodo por def no pudo
    #haber quedado desbalanceado
    self.balancea(ultimo_nodo_insertado.padre)
      
  def borra(self, dato:float):
    """
    Metodo que borra un nodo en el arbol y lo balancea, si es que se desbalanceo
    inputs:
        dato: elemento a borrar
    outputs:
        ---

    >>>ArbolAVL(13).borra(13) 
    <  >
    """
    #Se borra el nodo usando el metodo de la clase padre
    borradopadre = self.borra_nodo(self.raiz, dato)
    
    #Se rebalancean los nodos que hayan quedado desbalanceados
    #Y se empieza desde el padre, por que nodo por def, no pudo
    #haber quedado desbalanceado
    self.balancea(borradopadre)
```

### Análisis

Indica cuál es el peor caso para este tipo de árboles y explica a qué orden de complejidad pertenecen las operaciones: eliminar árbol, buscar, insertar.

*Respecto a la complejidad en número de operaciones*

**Peor caso**

Notemos que la propiedad particular (de balanceo) en este tipo de árbol binario; evita que se cree un árbol degenerado (el cual es el peor caso en un ABB).

Entonces, el peor caso para este tipo de árbol debe ser aquel en que el elemento buscado se encuentra en la peor posición posible. Esto es, en una hoja de mayor profundidad. No obstante, para este escenario, la característica de orden de estos árboles permite que, en cada consulta a cada nodo, se descarte a la mitad de nodos faltantes por comparar. 

Por lo tanto, se alcanza un orden de complejidad logarítmico. Es decir: $O(log_2 n)$.


**Eliminar árbol**

Al igual que en los árboles binarios de búsqueda, esta operación es de orden constante: $O(1)$, pues una vez que se elimina al nodo raíz del árbol, se libera todo el espacio en memoria asociado a éste. De manera que, no quedan objetos con los cuales realizar operaciones.

Por lo tanto, con una única operación se elimina todo el árbol.

**Buscar**

Dado que sus nodos se encuentran correctamente ordenados y blanceados, al buscar un nodo no se deben recorrer todos los nodos pertenecientes al arbol, solo una pequeñas cantidad dado que se van descartando al avanzar por las hojas. 

Por lo tanto, el número de comparaciones a realizar para encontrar un elemento pertenece al orden de complejidad logaritimica, es decir:  $O(log_2n) $  (donde  $n$  denota el número de nodos que posee el AAVL).


**Insertar**

Al insertar un elemento se sigue una lógica muy similar que al buscarlo, se realiza un número de comparaciones para asegurar que es insertado en el orden correspodiente, sin embargo por como se construye el arbol, se descartan varios casos de comparación y así, el número de comparaciones oara poder insertar un elemento en un arbol AVL pertenece al orden de complejidad logarítimica, es decir $O(log2n) $ (donde  $n$  denota el número de nodos que posee el AAVL).

```python
# Se obtiene el archivo a analizar
!wget https://raw.githubusercontent.com/jugernaut/ManejoDatos/desarrollo/utils/inteligencia.txt
```

```python
#Se declara el archivo a utilizar
texto = 'inteligencia.txt' 
# Lista auxuliar
conteo = [] 

try: 
  # se abre el archivo
  with open(texto) as text:  
    # se crea una variable que almacene el contenido del archivo
    contenido = text.read()  

    # lista de caracteres a eliminar 
    caracteresNoDeseados = ",;:.!\"\'\¿\?\(\)" 
    # se convierte todo el archivo a letras minúsculas
    contenido = contenido.lower() 

  for caracter in caracteresNoDeseados: 
    # se eliminan los caracteres no deseados
    contenido = contenido.replace(caracter, "") 

# establecemos un error de salida si el archivo no se encuentra
except FileNotFoundError: 
  print(f'Parece que el archivo {texto} no existe') 

else: 
  # se crea una variable que almacene solo palabras del archivo
  palabras = contenido.split() 

  for palabra in palabras: 
    # se añade cada palabra con su respectivo conteo en forma de diccionario a una lista
    if palabra not in conteo: 
      conteo. append( {palabra: palabras.count(palabra)})

  # se crea un diccionario auxiliar 
  conteo_palabras = {} 

  for i in conteo: 
    # se unen los diccionarios almacenados anteriormente en la lista para formar uno solo
    conteo_palabras.update(i) 
  
  # Se imprime en pantalla el resultado obtenido
  print('Diccionario de las palabras del texto con su respectiva ocurrencia. ')
  print(conteo_palabras)
```
